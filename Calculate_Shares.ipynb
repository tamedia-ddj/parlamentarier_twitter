## Anteile der Beziehungen berechnen

```python
import pandas as pd
```

```python
followers_complete = pd.read_pickle("data/followers_complete.pkl")
```

```python
followers_complete
```

```python
def calc_own_share(parties):
    own_share = []
    for party in parties:
        n_connections = len(followers_complete[(followers_complete.follower_Party == party) &
                                               (followers_complete.user_Party == party)])
        n_members = len(followers_complete[(followers_complete.follower_Party == party)].follower_ID.unique())
        
        connections_per_member = n_connections / n_members
        
        result = 100 / n_members * connections_per_member
        
        own_share.append(result)
        
    return own_share

def calc_foreign_share(parties):
    foreign_share = []
    for party in parties:
        n_connections = len(followers_complete[(followers_complete.follower_Party == party) &
                                               (followers_complete.user_Party != party)])
        n_members = len(followers_complete[(followers_complete.follower_Party == party)].follower_ID.unique())
        n_foreign_account = len(followers_complete[(followers_complete.follower_Party != party)].follower_ID.unique())
        
        connections_per_member = n_connections / n_members

        result = 100 / n_foreign_account * connections_per_member
        
        foreign_share.append(result)
        
    return foreign_share
    
```

```python
follower_party_data = followers_complete[["follower_Party", "follower_ID"]].groupby(["follower_Party", "follower_ID"]).count().reset_index().groupby(["follower_Party"]).count().reset_index()
```

```python
follower_party_data["own_share"] = calc_own_share(follower_party_data.follower_Party)
```

```python
follower_party_data["foreign_share"] = calc_foreign_share(follower_party_data.follower_Party)
```

```python
follower_party_data.columns = ["follower_Party", "n_Accounts", "own_share", "foreign_share"]
```

```python
follower_party_data["share_diff"] = follower_party_data["own_share"] - follower_party_data["foreign_share"]
```

```python
follower_party_data.sort_values(by="share_diff", ascending=False)
```
